```python
#df_avg_saleprice_midwest, Quarterly data
df_avg_saleprice_midwest = pd.read_csv("C:\\Home New\\Avg Sales price houses sold in midwest.csv", names = ["DATE", "Avg_sale_priceHS_Midwest"], skiprows = 1)
print(df_avg_saleprice_midwest.shape)
df_avg_saleprice_midwest.head()
```

```python
#Current_employement_statistics, monthly data
df_emp_statistics = pd.read_csv("C:\\Home New\\Current_employement_statistics.csv", names = ["DATE", "Working_population"], skiprows = 1)
print(df_emp_statistics.shape)
df_emp_statistics.head()
```

```python
#Home ownership rate, Quartely data
df_home_owner = pd.read_csv("C:\\Home New\\Home Ownership rate in US.csv", names = ["DATE", "Homeownership_rate"], skiprows = 1)
print(df_home_owner.shape)
df_home_owner.head()
```

```python
#Median Sales price, Qurterly data
df_Median_Sales_price = pd.read_csv("C:\\Home New\\Median sales price for houses sold in US.csv", names = ["DATE", "Median_Sales_price_Home"], skiprows = 1)
print(df_Median_Sales_price.shape)
df_Median_Sales_price.head()
```

```python
#Median Household income, Yearly data
df_Median_Household_income = pd.read_csv("C:\\Home New\\Median_household_income.csv", names = ["DATE", "Median_Household_Income"], skiprows = 1)
print(df_Median_Household_income.shape)
df_Median_Household_income.head()
```

```python
#Mortgage percent, Yearly data
df_Mortgage_percent = pd.read_csv("C:\\Home New\\MORTGAGE30US percent.csv", names = ["DATE", "Mortgage_percent"], skiprows = 1)
print(df_Mortgage_percent.shape)
df_Mortgage_percent.head()
```

```python
#Vacant houses inventory , Quarterly data
df_vacant_inventory = pd.read_csv("C:\\Home New\\Vacant houses inventory units.csv", names = ["DATE", "Vacant Houses Inventory"], skiprows = 1)
print(df_vacant_inventory.shape)
df_vacant_inventory.head()
```

```python
#Occupied Houses by owners
df_occupied = pd.read_csv("C:\\Home New\\Occupied Housing units.csv", names = ["DATE", "Occupied by owners"], skiprows = 1)
print(df_occupied.shape)
df_occupied.head()
```

```python
#Employement -Population ratio
df_emp_pop_ratio = pd.read_csv("C:\\Home New\\Employement population ratio.csv", names = ["DATE", "Emp_Population_ratio"], skiprows = 1)
print(df_emp_pop_ratio.shape)
df_emp_pop_ratio.head()
```

```python
#US Population
df_population= pd.read_csv("C:\\Home New\\US Population.csv", names = ["DATE", "Population"], skiprows = 1)
print(df_population.shape)
df_population.head()
```

```python
#Housing Inventory from 2016
df_Hou_inv= pd.read_csv("C:\\Home New\\Housing Inventory.csv", names = ["DATE", "Total_Housing_inventory_2016"], skiprows = 1)
print(df_Hou_inv.shape)
df_Hou_inv.head()
```

```python
#Let us merge data

df= pd.DataFrame()
df_bymonth= [df_cs,df_emp_statistics,df_unemp,df_gdp,df_Fed_rate,df_Const_price_index,df_income,df_new_hs,
             df_avg_saleprice_midwest,df_avg_saleprice_northeast,df_home_owner,df_consumer_goods_Avg_price_index,
            df_Mortgage_percent,df_vacant_inventory,df_occupied,df_emp_pop_ratio,df_population,df_Hou_inv,df_Median_Sales_price]
for df1 in df_bymonth:
    df1["DATE"] = pd.to_datetime(df1["DATE"])
    df1 = df1.set_index("DATE")
    df = pd.concat([df,df1], axis = 1)
print(df.shape)
df.head()
```

```python
# Filling missing values in the Per_Capita_GDP column using linear interpolation
df["GDP_per_capita"] = df["GDP_per_capita"].interpolate()
```

```python
df["Vacant Houses Inventory"] = df["Vacant Houses Inventory"].interpolate()
```

```python
df["Occupied by owners"] = df["Occupied by owners"].interpolate()
```

```python
df["Median_Sales_price_Home"] = df["Median_Sales_price_Home"].interpolate()
```
